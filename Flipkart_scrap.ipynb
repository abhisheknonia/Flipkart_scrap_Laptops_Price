# Function for above logic

```python
import pandas as pd
def Flipkart(soup):
    Product_Name=[i.text.strip() for i in soup.select('div._4rR01T')]
    off_Price1=[''.join(i.text.strip()[0:9].split('₹')[1].split(',')) for i in soup.select('div._25b18c')]
    Rating=[i.text.strip() for i in soup.select('div._3LWZlK')]
    Total_Rating1 = [i.text.split(' ')[0] for i in soup.select('span._2_R_DZ')]
    total_reviews1=[i.text.split('Ratings\xa0&\xa0')[1].split(' ')[0] for i in soup.select('span._2_R_DZ')]
    List_specification = [i.text.strip() for i in soup.select('ul._1xgFaf')]
    return pd.DataFrame({'Product_Name':Product_Name,'Price in Rs.':off_Price1,'Rating out of 5':Rating,'Total_Rating': Total_Rating1,'Total_Reviews':  total_reviews1,'List_specification': List_specification})
```

```python
df=Flipkart(soup)
```

```python
tables=[]
for i in range(1,3):
    urls=url.format(i)
    browser.get(url)
    html=browser.page_source
    soup=BeautifulSoup(html,'lxml')
    tables.append(Flipkart(soup))
```

```python
Laptops=pd.concat(tables,axis=0)
Laptops.head()
```
